```python
import pandas as pd
df = pd.read_csv('titanic_train.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(['PassengerId', 'Name', 'SibSp','Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
target = df.Survived
features = df.drop('Survived', axis='columns')
```

```python
# ML models cannot handle text
dummies = pd.get_dummies(features.Sex)
dummies.head()
```

```python
features = pd.concat([features, dummies], axis='columns')
```

```python
features.head()
```

```python
features.drop('Sex', axis='columns', inplace=True)
features.head()
```

```python
features.columns[features.isna().any()]
```

```python
features.Age[:10]
```

```python
features.Age = features.Age.fillna(features.Age.mean())
```

```python
features.columns[features.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(features)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
