> **Question 2**<br/>
>learn the attributes

- **survived** <br/> True/False (0/1) whether or not the passenger survived or died from the sinking

- **pclass** <br/> The accomodation type of the passenger. This corresponds to economic status.<br/> pclass 1 > pclass 2 > pclass 3
- **sex** <br/> Male/Female
- **age** <br/> Years old age of passenger <br/> Children less than 1 are given a decimal value
- **sibsp** <br/> Siblings and spouces on board
- **parch** <br/> Parents and Children on board
- **fare** <br/> amount paid for fare
- **embarked** <br/> Port the passenger boarded/embarked onto the chip<br/> C is Cherebourg, Q is Queenstown, S is Southhampton
- **class** <br/> The ordinal version of the pclass.
- **who** <br/> female/male if adult (age >= 16). Child if child (age <= 15)
- **adult_male** <br/> age >= 16 and is a male
- **deck** <br/> The deck the passenger stayed on
- **embark_town** <br/>Port the passenger boarded/embarked represented as full name instead of C,Q,S representation
- **alive** <br/> Wether the passenger lived or died, same as surived but as a yes or no statment
- **alone** <br/> No siblings, spouses, parents, or children on board
