```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
#Top 5 tips
tips.head()
```

```python
#Histogram to demonstrate distribution of data
sns.displot(tips['total_bill'])
```

```python
#Show the curve of the histogram
sns.displot(tips['total_bill'], kde=True)
```

```python
#Best graphs to compare each two data fields
sns.pairplot(tips, hue = 'sex',palette='coolwarm')
```

```python
#Total bill by sexes
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.boxenplot(x='day',y='total_bill', data=tips, palette='rainbow')
```

```python
#Box plot to show locality and spread 
sns.boxenplot(x='day',y='total_bill', data=tips, hue='smoker', palette='coolwarm')
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
#Show correlation of all the data fields
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```
